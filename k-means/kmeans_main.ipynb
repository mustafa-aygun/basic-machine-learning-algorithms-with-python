```python
import numpy as np
from kmeans import kmeans
from kmeans import assign_clusters
import matplotlib.pyplot as plt 
```

```python
dataset1 = np.load('kmeans/dataset1.npy')
dataset2 = np.load('kmeans/dataset2.npy')
dataset3 = np.load('kmeans/dataset3.npy')
dataset4 = np.load('kmeans/dataset4.npy')
```

```python
def random_centroids(k,data):
   initial_cluster_centers = np.random.uniform(data.min(), data.max(), size=(k, 2))
   return initial_cluster_centers

all_objective_functions = []
all_cluster_centers = []
for i in range(10):
   one_row = []
   one_row_obj_function = []
   for j in range (10):
      initial_cluster_centers = random_centroids(i+1,dataset1)
      cluster_centers, objective_function = kmeans(dataset1,initial_cluster_centers)
      one_row.append(cluster_centers)
      one_row_obj_function.append(objective_function)
   all_cluster_centers.append(one_row)
   all_objective_functions.append(one_row_obj_function)

np.save("dataset1_cluster_centers.npy",np.array(all_cluster_centers,dtype=object))
a = np.load("dataset1_cluster_centers.npy",allow_pickle=True)
#Plot graphs 
min_objective_functions = np.min(all_objective_functions,axis=1)
plotDataSet1 = plt.figure(1)
plt.plot(range(1,11),min_objective_functions)
plt.xticks(range(1,11))
plt.title('K-Means/DataSet1')
plt.xlabel('K-value')
plt.ylabel('Objective Function')
plt.show()
```

```python
all_objective_functions = np.array(all_objective_functions)
indexs = all_objective_functions.argmin(axis=1)
last_cluster_centers = [] 
a = np.array(a)
for i in range(10):
    last_cluster_centers.append(a[i][indexs[i]])

belongs_to = assign_clusters(dataset1,last_cluster_centers[1])
draw_cluster_centers = last_cluster_centers[1]

plt.scatter(dataset1[belongs_to == 0,0],dataset1[belongs_to == 0,1],color='cyan')
plt.scatter(dataset1[belongs_to == 1,0],dataset1[belongs_to == 1,1],color='green')
plt.scatter(draw_cluster_centers[:,0],draw_cluster_centers[:,1],marker='^',s=150,color='black')
plt.title('Data Distrubition and Centers/Dataset1')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.show()
```
